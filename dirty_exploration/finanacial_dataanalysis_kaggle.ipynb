```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
import os
import shutil
from pathlib import Path
import pandas as pd
```

```python
notebook_dir = Path.cwd()
target_dir = notebook_dir / "data"
target_dir.mkdir(exist_ok=True)

dataset_path = kagglehub.dataset_download("sanjanapatil7/largest-companies-in-usa-by-revenue")
print(f"Dataset downloaded to cache: {dataset_path}")
```

```python
files = os.listdir(dataset_path)
print(f"\nAvailable files: {dataset_path}")
```

```python
print(f"\nFiles in cache directory:")
files = [f for f in os.listdir(dataset_path) if os.path.isfile(os.path.join(dataset_path, f))]
for file in files:
    print(f"  - {file}")
```

```python
for file in files:
    src = os.path.join(dataset_path, file)
    dst = target_dir / file
    shutil.copy2(src, dst)
```

```python
if files:
    file_name = files[0]
    file_path = target_dir / file_name
    
    if file_name.endswith('.csv'):
        df = pd.read_csv(file_path)
```
